```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../data/processed/cleaned_claims.csv")
```

```python
#Convert target variable for easier analysis
df['fraud_flag'] = df['fraud_reported'].map({'Y': 1, 'N': 0})
```

FRAUD DISTRIBUTION

```python
df['fraud_reported'].value_counts(normalize=True).plot(
    kind='bar',
    title='Fraud vs Non-Fraud Claims Distribution'
)
plt.ylabel('Proportion')
plt.show()
```

CUSTOMER BEHAVIOUR ANALYSIS

```python
# Age vs Fraud
sns.boxplot(x='fraud_reported', y='age', data=df)
plt.title('Age Distribution by Fraud Status')
plt.show()
```

```python
# Customer Tenure vs Fraud
sns.boxplot(x='fraud_reported', y='months_as_customer', data=df)
plt.title('Customer Tenure vs Fraud')
plt.show()
```

INCIDENT-LEVEL ANALYSIS

```python
# Incident Type vs Fraud
plt.figure(figsize=(10,4))
sns.countplot(x='incident_type', hue='fraud_reported', data=df)
plt.xticks(rotation=45)
plt.title('Incident Type vs Fraud')
plt.show()
```

```python
# Authorities Contacted vs Fraud
plt.figure(figsize=(10,4))
sns.countplot(x='authorities_contacted', hue='fraud_reported', data=df)
plt.xticks(rotation=45)
plt.title('Authorities Contacted vs Fraud')
plt.show()
```

FINANCIAL SIGNALS

```python
# Premium vs Fraud
sns.boxplot(x='fraud_reported', y='policy_annual_premium', data=df)
plt.title('Policy Premium vs Fraud')
plt.show()
```

```python
# Claim Amount vs Fraud
sns.boxplot(x='fraud_reported', y='total_claim_amount', data=df)
plt.title('Claim Amount vs Fraud')
plt.show()
```

Correlation Analysis (Numerical)

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.select_dtypes(include='number').corr(), cmap='coolwarm')
plt.title('Feature Correlation Heatmap')
plt.show()
```

Key Insights from EDA:

1. Fraudulent claims form a minority, requiring precision-focused models.

2. Newer customers show higher fraud tendency.

3. Certain incident and collision types are more fraud-prone.

4. Lack of authority involvement correlates with fraud.

5. Fraudulent claims typically involve higher claim amounts.
